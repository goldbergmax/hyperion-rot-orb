```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# chirikov standard map
def chirikov(x, p, K):
    return (K*np.sin(x), x + p + K*np.sin(x))

N_iter = 1000
N_samp = 144
K = 0.971635
res = np.zeros((N_iter, N_samp, 2))
res[0,:,:] = np.array(np.meshgrid(np.linspace(0, 2*np.pi, int(np.sqrt(N_samp))), np.linspace(0, 2*np.pi, int(np.sqrt(N_samp))))).T.reshape(-1,2)
# iterate map
for i in range(1,N_iter):
    p = res[i-1,:,0] + K*np.sin(res[i-1,:,1])
    x = res[i-1,:,1] + p
    res[i,:,:] = np.array([p, x]).T
# res = np.remainder(res, 2*np.pi)
```

```python
plt.scatter(np.remainder(res[:,:,1], 2*np.pi), np.remainder(res[:,:,0], 2*np.pi), s=0.03)
plt.gca().set_aspect('equal', adjustable='box')
```
